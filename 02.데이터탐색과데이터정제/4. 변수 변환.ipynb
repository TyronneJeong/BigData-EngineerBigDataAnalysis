# 4. 변수 변환

이번에는 데이터 변환 과정을 통해 이상치 문제를 해결 하는 방법을 알아 보겠습니다. 변수변환의 대표적인 방법으로는 **log변환, 제곱근 변환**이 있습니다. 두 방법 모두 <u>데이터 단위를 크게 줄여서 큰 값이 주는 영향을 대폭 낮추는 것에 의미</u>가 있습니다. 손쉽게 이상치를 해결할 수 있다는 장점이 있는 반면, **단위가 변화기 때문에 결과 해석이 어렵다는 단점도 동반** 됩니다.

## 4.1 로그(Log)변환

로그 변환은 가장 많이 활용되는 변수변환 방법입니다. 변수 변환은 판다스에서는 지원되지 않으며 numpy 에서만 작업 가능하기 때문에. 넘파이를 불러와야 합니다.

```python
# numpy 를 통한 로그변환
import pandas as pd
data = pd.read_csv('./extrafiles/Ex_CEOSalary.csv', encoding='utf-8')

import numpy as np
data['log_salary'] = np.log(data['salary'])
data['log_sales'] = np.log(data['sales'])
data['log_roe'] = np.log(data['roe'])
data.head()
```

원본 데이터와 log 변환 데이터를 나란히 히스토그램으로 보면, 이상치의 값이 중앙으로 이동한 효과를 확인 할 수 있다.

```python
# log 변환을 취한 변수들은 전반적으로 오른쪽으로 이동 하였다.
import matplotlib.pyplot as plt
data.hist(bins=50, figsize=(20, 15))
```

```python
# 상관계수 조회
data.corr()
```

상관관계를 보면 log 를 취하지 않은 원본 데이터의 경우 salary 와 sales 간의 상관관계가 0.11 정도로 낮은 상관관계를 보인다.
하지만 **log 변환을 취한 이후에는 0.45, 0.16 등으로 비교적 뚜렷한 상관관계를 나타나게 되었다.**
이로써 로그변환이후 왜곡이 큰 값들이 상당수준 해소된것을 확인 할 수 있다.  

<u>※ 단일항 하나만 로그값을 취하는 것이 아니라 로그 변환을 취한 값들 끼리 상관관계를 확인하는 식이다.</u>
