# 3. 이상치 처리

이상치(Outlier) 는 분석 전 반드시 해결 해야하는 문제 입니다. 정상적인 범위를 지나치게 벗어난 작은 값 또는 큰 값은 전체 분석 결과에 큰 영향을 미칠수 있으므로 보통의 경우 이상치는 제거 하도록 합니다.

이상치를 시각적으로 파악 하기 위해서는 **상자수염도표(box-whisker plot)** 을 활용합니다.  
opt) 옵션 타입으로는 axes, dict, both 세가지가 있다.

```python
# box plot
import pandas as pd
data = pd.read_csv('./extrafiles/Ex_CEOSalary.csv', encoding='utf-8')

a = data.boxplot(column='salary', return_type='both')


print(a)
```

```python
data.boxplot(column='sales', return_type='both')
```

```python
data.boxplot(column='roe', return_type='both')
```

## 3.1 salary 변수 이상치 처리

이상치를 판단하는 명쾌한 기준은 없습니다. 분석자의 주관적인 판단에 따라 제거하는 것이 일반적입니다.  
**IQR(Inter Quartile Rane)-사분위편차**를 기준으로 이상치를 제거하는 과정을 진행 해 봅시다.

진행과정
**우선 1산분위수(Q1)와 3사분위수(Q3) 값을 구하여 사분위편차(Q3-Q1)를 알아냅니다.**

```python
# 데이터 프레임에서 사분위 수를 산출 하여 저장한다. quantile
Q1_salary = data['salary'].quantile(q=0.25)
Q3_salary = data['salary'].quantile(q=0.75)

print("1사분위수 : {}".format(Q1_salary))
print("3사분위수 : {}".format(Q3_salary))

IQR_salary = Q3_salary - Q1_salary
IQR_salary
print(f"사분위편차 : {IQR_salary}") # 3.6 부터 추가된 포맷 스트링
```

이제 IQR 사분위 편차를 알아내었으니 **IQR 의 1.5배 큰 값과작은 값을 이상치의 기준 값**으로 정하고 제거해 봅시다.

```python
# 이상치 제거 후 히스토 그램 생성
data_IQR = data[(data['salary'] < Q3_salary+IQR_salary*1.5) & # 3분위수에 IQR 1.5배 이하의 값과
                (data['salary'] > Q1_salary-IQR_salary*1.5)]  # 1분위수에 IQR 1.5배 이상인 값만 출력
data_IQR['salary'].hist()
                
```

```python
# IQR 사분위수 기준으로 salary 의 이상치를 제거한 데이터 프레임의 히스토 그램
data_IQR.hist(bins=50, figsize=(20, 15))
```

사분위 편차의 1.5배를 상하한으로 정하여 이상치 제거를 실행한 <u>salary 의 경우 극단 값이 제거 되어 완만한 정규분포의 곡선 형태로 보여지게 되었다.</u> 책에서는 ROE 의 분포곡선도 좋아졌다고는 하나 차이가 거의 없다.

```python
# 각 변수간 상관관계에 대해서도 알아보자
data_IQR.corr() 
```

이상치가 제거 된 이후 각 변수간 상관관계가 더욱 높아짐을 확인 할 수 있다.

```python
# 산점도를 통해서 변화된 내용을 확인해 보자
import matplotlib.pyplot as plt
plt.scatter(data_IQR['sales'], data_IQR['salary'], alpha=0.5)
plt.show()
```

여전히 이상치는 확인이 된다. 이는 sales 열의 이상치이므로 다음 단계에서 제거하도록 한다.

```python
plt.scatter(data_IQR['roe'], data_IQR['salary'], alpha=0.5)
plt.show()
```

## 3.2 sales 변수 이상치 처리

이번에는 sales 변수의 이상치를 제거하고 정상 범위의 값들만 새로운 데이터 프레임으로 받아 보자.
sales 변수 역시 IQR 사분위편차 기준으로 1분위수에서 IQR 1.5 배 이하의 값 이상. 3분위 수에서 IQR 1.5배 값 이하인 값 들을 정상치로 보고 이를 추출하여 보자.



```python
# 데이터 프레임에서 사분위 수를 산출 하여 저장한다. quantile
Q1_sales = data['sales'].quantile(q=0.25)
Q3_sales = data['sales'].quantile(q=0.75)

print("1사분위수 : {}".format(Q1_sales))
print("3사분위수 : {}".format(Q3_sales))

IQR_sales = Q3_sales - Q1_sales
IQR_sales
print(f"사분위편차 : {IQR_sales}") # 3.6 부터 추가된 포맷 스트링
```

```python
# sales와 salary 두 변수 모두 이상치가 제거된 데이터 프레임을 생성 한다.
data_IQR = data[
    (data['sales'] < Q3_sales  + IQR_sales *1.5) &
    (data['sales'] > Q1_sales  - IQR_sales *1.5) &
    (data['salary']< Q3_salary + IQR_salary*1.5) &
    (data['salary']> Q1_salary - IQR_salary*1.5) 
]
data_IQR['sales'].hist()
```

```python
# data_IQR 변화된 히스토그램 확인
data_IQR.hist(bins=50, figsize=(20, 15)) # bins 는 분할되는 구간의 개수, figsize 는 그려지는 그림의 크기
```

```python
# 변수들간 상관 관계가 개선 되었다.
data_IQR.corr()
```

정리하자면 이상치는 머신러닝 분석 전 반드시 해결해야할 자료입니다. 가장 확실한 방법은 너무 큰 값, 너무 작은  값을 제거 하는 것이며, 그 기준은 분석자마다 상황마다 다릅니다.
