# Pandas 주요 함수 정리

```python
import numpy as np
import pandas as pd

df = pd.read_csv('./extrafiles/Train.csv')
print(df.columns)
df
```

```python
# 데이터 프레임에 신규 컬럼을 생성하고 모두 1을 대입한다.ㄴ
df['Gender_Value'] = 1

# map 함수를 통해 매핑을 한다.
df['Gender_Value'] = df['Gender'].map({'F':0, 'M':1})
df['Reached.on.Time_Y.N'].map({1:'NO', 0:'YES'})
```

```python
# LabelDecoder
df['Warehouse_block'].value_counts(0).sort_index()
print(df['Warehouse_block'])

# A, B, C, D, F -> 1, 2, 3, 4, 6
mapping = {'A':1, 'B':2, 'C':3, 'D':4, 'F':6}
df['WB_VAL'] = df['Warehouse_block'].map(mapping)
df['WB_VAL']

# 다시 replace
df.replace({'WB_VAL':{1:11}})
df['WB_VAL'].replace({2:22}, inplace=True)
df['WB_VAL']
```

```python
# masking을 Lambda로
df['WB_CODE'] = (lambda x : x > 2)(df['WB_VAL'])
df['WB_CODE']
```

## 문제3
- 데이터셋(basic1.csv)의 'age'컬럼의 이상치를 더하시오!
- 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('./extrafiles/basic1.csv')
```

```python
# mask = (lambda x: (x > x.mean() + x.std()*1.5) 
#  | (x < x.mean() - x.std()*1.5))(df['age'])
df.loc[(lambda x: (x > x.mean() + x.std()*1.5) | 
        (x < x.mean() - x.std()*1.5))(df['age']), 'age'].sum()
```

```python
df_temp = df.groupby(by='f4').sum().sort_values('f4')
df_temp
```

```python
# drop
df_temp.drop(['INTJ'], axis=0) # 행 삭제
df_temp.drop(['f1'], axis=1) # 열 삭제
# df_temp['f5'].idxmax()

df_temp.value_counts()
df_temp.nunique()
```
