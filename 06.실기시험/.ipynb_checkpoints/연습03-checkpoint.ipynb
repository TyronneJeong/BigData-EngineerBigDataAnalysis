```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcwd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이브 경로(C: 등) 접근 불가

# 데이터 파일 읽기 예제
import pandas as pd
X_test = pd.read_csv("data/X_test.csv")
X_train = pd.read_csv("data/X_train.csv")
y_train = pd.read_csv("data/y_train.csv")

# 사용자 코딩

# 답안 제출 참고
# 아래 코드 예측변수와 수험번호를 개인별로 변경하여 활용
# pd.DataFrame({'cust_id': X_test.cust_id, 'gender': pred}).to_csv('003000000.csv', index=False)

# 시스템 셋팅
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 30)
pd.set_option('display.width', 2000)

import warnings
warnings.filterwarnings("ignore")

# EDA
# print(X_train.isna().sum())
# print(X_test.isna().sum())
# print(X_train.info())
# print(X_train.describe())


# 이상치/결측치 제거
X_train.환불금액.fillna(0, inplace=True)
X_test.환불금액.fillna(0, inplace=True)

X_test_id = X_test[['cust_id']]

# 불필요 컬럼 제거 : cust_id 제거
X_train.drop(['cust_id', '주구매지점'], axis=1,  inplace=True)
X_test.drop(['cust_id', '주구매지점'], axis=1,  inplace=True)

# 정규화
# 범주형 : 라벨링 또는 원핫 
# 연속형 : 스케일링(정규화)

# 연속형/범주형 데이터 분리
X_train_num = X_train[['총구매액', '최대구매액', '환불금액', '내점일수', '내점당구매건수', '주말방문비율', '구매주기']]
X_train_cat = X_train[['주구매상품']]
X_test_num = X_test[['총구매액', '최대구매액', '환불금액', '내점일수', '내점당구매건수', '주말방문비율', '구매주기']]
X_test_cat = X_test[['주구매상품']]

# 원핫인코딩
X_train_dummy = pd.get_dummies(X_train_cat)
X_test_dummy = pd.get_dummies(X_test_cat)

# 카테고리에 X_train_cat / X_test_cat 비교하여 없는 컬럼 채워 넣기
sample = [v for v in X_train_dummy.columns if v not in X_test_dummy.columns]
X_test_dummy[sample] = 0 # 주구매상품_소형가전 - 미존재 카테고리값 충전


# Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train_num)
X_train_scaled = scaler.transform(X_train_num)
X_train_scaled = pd.DataFrame(X_train_scaled, index=X_train_num.index, columns=X_train_num.columns)
X_test_scaled = scaler.transform(X_test_num)
X_test_scaled = pd.DataFrame(X_test_scaled, index=X_test_num.index, columns=X_test_num.columns)

# 데이터셋 합치기
X_train = pd.concat([X_train_scaled, X_train_dummy], axis=1)
X_test = pd.concat([X_test_scaled, X_test_dummy], axis=1)
y_train = y_train[['gender']]

# 모델 피팅
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
pred_train = model.predict_proba(X_train)
print("정확도 : ", model.score(X_train, y_train))

# ROC-AUC_SCORE 확인
from sklearn.metrics import roc_auc_score
print("ROC AUC Score : >> ", roc_auc_score(y_train, pred_train[:, 1]))

# 파일 저장
pred_test = model.predict_proba(X_test)
result = pd.DataFrame({'custid':X_test_id['cust_id'], 'gender':pred_test[:, 1]})
print(result)

result.to_csv('00001.csv', index=False)

```
