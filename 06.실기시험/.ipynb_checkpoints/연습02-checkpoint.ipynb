```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcwd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이브 경로(C: 등) 접근 불가

# 데이터 파일 읽기 예제
import pandas as pd
X_test = pd.read_csv("data/X_test.csv")
X_train = pd.read_csv("data/X_train.csv")
y_train = pd.read_csv("data/y_train.csv")

# 사용자 코딩
# 출력 라인 조정
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 10)
pd.set_option('display.width', 1000)

df = pd.concat([X_train, X_test], axis=0)

# 데이터 정제 - feature engineering, 이상치, 결측치 제거
df.fillna(0, inplace=True)
df.drop(['총구매액', '환불금액'], axis=1, inplace=True)
df = df[df['내점일수']<30]
df = df[df['최대구매액']<1.5e+7]

X_num = df[['최대구매액', '내점일수', '내점당구매건수', '주말방문비율', '구매주기']]
X_cat = df[['cust_id', '주구매상품']]

# print(X_num.shape)
# print(X_num.describe())


# 데이터 정규화
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_num)
X_num_scaled = scaler.transform(X_num)
X_num_scaled = pd.DataFrame(X_num_scaled, index=X_num.index, columns=X_num.columns)
X_cat_dummies = pd.get_dummies(X_cat)
X_scaled = pd.concat([X_cat_dummies, X_num_scaled], axis=1)

# 트레인-테스트 데이터 분리
X_scaled_train = X_scaled[X_scaled['cust_id']<3500]
X_scaled_test = X_scaled[X_scaled['cust_id']>=3500]

# 동일 cust_id 데이터 추출
y_train = y_train.loc[y_train['cust_id'].isin(X_scaled_train['cust_id'])]
# y_train = y_train[['gender']]
# 모델 적용
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_scaled_train, y_train)
pred_train = model.predict(X_scaled_train)

# print(X_scaled_train)
print(y_train)

# print(model.score(X_scaled_train, y_train))

# 모델 적용
# from sklearn.ensemble import AdaBoostRegressor
# model = AdaBoostRegressor(random_state=0, n_estimators=100)
# model.fit(X_scaled_train, y_train)
# pred_train = model.predict(X_scaled_train)
# print(model.score(X_scaled_train, y_train))

# 데이터 검증 - 분류는 confusionMatrix, 회귀는 R-square, RMSE

# RMSE (Root Mean Squared Error)
from sklearn.metrics import mean_squared_error
import numpy as np

MSE_train = mean_squared_error(y_train, pred_train)
print(np.sqrt(MSE_train))


# 하이퍼 파라미터 최적화








# 답안 제출 참고
# 아래 코드 예측변수와 수험번호를 개인별로 변경하여 활용
# pd.DataFrame({'cust_id': X_test.cust_id, 'gender': pred}).to_csv('003000000.csv', index=False)
```
