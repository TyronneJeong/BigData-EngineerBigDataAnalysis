

# 1. 빅데이터 분석기사 실기 시험 특징

## 1.1 시험 환경 특징

#### 빅데이터 분석기사 실기시험에 대한 안내를 보면 몇가지 특징이 있다.

---

### 실기시험 응시환경에 대해 충분히 확인하고 준비하여 주시기 바랍니다.

**■ 제약사항**
1. 라인별 실행, 그래프 기능, 단축키, 자동완성 기능 <u>미제공</u>
2. <u>코드 실행 시간은 1분으로 제한  </u>
3. 패키지는 추가 설치 할 수 없으며, 제공된 패키지만 이용가능  

**■  실기 응시환경 내에서 help 함수 이용 가능**

**■  답안제출**
1. 단답형
 - 답안 작성후 "제출"버튼 클릭, 답안은 영문, 한글 모두 가능
2. 작업형 제1유형
 - 마지막 라인에 print(변수명)명령어로 출력하는 코드를 제출
3. 작업형 제2유형
 - 수험번호.csv 파일을 생성하는 코드를 작성 하여 실행 한 후 제출


※ 답안은 중복 제출 가능하며, 마지막 제출 답안이 채점 대상  
※ 작업형의 경우 임의의 값을 대입하여 출력하는 경우 0점 처리  
※ 작업형 제2유형채점은 제출된 csv 파일의 평가지표에 따라 구간별 점수를 획득.  
   (예. 지표가 0.5~0.6 일경우 10 점 등.)

---

첫 번째 제약 사항을 보며느 라인별 실행, 그래프 기능, 단축키, 자동완성 기능이 제공되지 않는다고 공지하고 있다. 먼저 라인별 실행이 되지 않는다는 것은 전체 분석 코드를 한번에 실행 하라응 의미이다. 따라서 응시자들은 분석을 하면서 각 단계별 코드 결과를 확인하기 보다 데이터를 불러오기 부터 머신 러닝 수행까지 전체 핵심 코드를 한번에 수행해야 한다. 단축키와 자동 완성 기능이 제공되지 않는다는 것은 코드를 외워서 응시해야 한다는 의미이므로, 제1 제약 사항을 정리하면 **'`전체 분석 수행코드를 암기하여 불러오기 부터 수행까지 진행 가능 해야 한다.`'** 이다.

두 번째 제약 사항은 코드 실행 시간을'1분으로 제한한다' 는 것이다. 이는 두 가지 측면을 유추 해 볼 수 있는데. 먼저 수만 건 이상의 대량 데이터는 시험에서 다루지 않는 것이고, 데이터가 많을수록, 변수가 많을수록 수행 시간이 오래 걸리므로 시험 출제기관에서는 최대 수천건 이내의 데이터로 시험을 보는것을 알 수 있다.

다음으로 그리드 서치(Grid Search) 랜덤 서치(Radnom Search) 와 같이 알고리즘의 최적 하이퍼 파라미터를 탐색 하는 과정은 시험 요건상 수행이 어려울 것으로 판단 된다. 따라서 제2 제약사항을 정리하면 **'`기본 알고리즘으로 최대 수천건의 데이터를 분석하는 시험을 출제 한다`'** 는 의미로 볼 수 있다.

세번째 제약 사항은 제공된 패키지만 이용 가능하다는 점이다. 2021년 3회 시험 현재 한국데이터산업 진흥원에서 공지한 파이썬 패키지 주요 라이브러리는 다음과 같다. 데이터 처리 관련 기본 패키지는 numpy 와 pandas 이다. 분석과 관련된 패키지는 sikit-learn, scipy, xgboos 가 있다. 반면 딥러닝이나 텍스트 분석과 관련된 패키지는 제공되지 않는다. 따라서 **'`기본 데이터 처리와 분석, 머신러닝 알고리즘 분석과 관련된 문제만 출제한다`'** 는 의미로 확볼 수 있다.

다만, beautifulsoup, selenium 등 웹 크롤링 패키지도 제공하여 특정 웹사이트에서 데이터를 수집하여 분석 하나는 문제으 출제도 가능하긴 하다. 다만 이는 난이도가 높아 본 시험의 수준에 맞지 않을것으로 예상 된다.

### 1.2 시험 문제 특징
#### 한국 데이터 산업 진흥원에서는 문제의 예시도 알려주고 있다.
<br/>
    
    
<center>[작업형 제1유형 : 데이터 처리 영역]</center>  

---

**작업형 제1유형 예시문제**

mtcars 데이터셋(mtcars.csv)의 qsec 컬럼을 최소최대 척도(Mix-Max Scal)로 변환한 후 0.5보다 큰 값을 가지는 레코드 수를 구하시오.

---

<center>[작업형 제2유형 : 모형 구축 및 평가 영역]</center>

**작업형 제2유형 예시문제**

아래는 백화점 고객의 1년간 구매 데이터 이다.


#### (가) 제공 데이터 목록  
① y-train.csv : 고객의 성별 데이터 (학습용).csv 형식의 파일  
② X_train.csv, X_test.csv : 고객의 상품구매 속성(학습용 및 평가용), csv 형식의 파일

#### (나) 데이터 형식 및 내용  
① y_train.csv (3,500명 데이터)  
- custid : 고객 ID  
- gender : 고객의 성별 (0:여자, 1남자)

② X_train.csv (3,500명 데이터), X_test.csv(2,482명 데이터)  


고객 3,500명에 대한 학습용 데이터(y_tarin.csv, X_train.csv)를 이용하여 성별예측 모형을 만든 후, 이를 평가용 데이터(X_test.csv)에 적용하여 얻은 2,482명 고객의 성별 예측값(남자일 확률)을 다믕과 같은 형식의 CSV 파일로 생성하시오.
<br/>  
<br/>  
<center> (제출한 모델의 성능은 ROC_AUC 평가지표에 따라 채점) </center>  
<br/>

**제출형식**

|custid.gender|
|------------:|
|   3500.0.267|
|   3501.0.578|
|   3502.0.885|
|          ...|

<br/>

**유의사항**

성능이 우수한 예측 모형을 구축하기 위해서는 **적절한 데이터 전처리, Feature Engineering 분류 알고리즘 사용, 초 매개변수 최적화, 모형 앙상블** 등이 수반 되어야 한다.

---
  
위 예제문제를 통해 문제 수준과 난이도를 가늠해보면 정확도를 높이는 등의 난이도 높은 문제보다는 **'데이터의 정제 및 변환을 할 줄 아는가'**, **'학습 및 테스트 데이터를 머신러닝 알고리즘을 적용하고 예측값을 csv 파일로 저장 할 수 있는가'** 등과 같이 전체적인 과정을 무난하게 수행하는 수준입니다.

따라서 첫째, **'파이썬으로 결측치와 이상 값 등 데이터 정체, 데이터 변환 및 생성 등 데이터 처리의 기본'**을 익히는 데에 시험 준비에 중점을 두어야 합니다.

둘째, 깊이 있는 알고리즘 이해와 고난이도 분석법보다 **'데이터 불러오기 -> 머신러닝알고리즘의 적용-> 학습모델로 테스트 데이터 예측 및 파일 저장 등 일련의 기본과정을 실수 없이 수행'** 할 수 있어야 합니다.

다만 시험의 기준과 난이도는 언제든 바뀔 수 있으므로 파이썬 데이터 다루기 등 기본에 충실 하면서 머신러닝 알고리즘을 다양하게 익혀두어야 합격에 의미가 있습니다.
