**1. Data Set 분할**  
학습 데이터를 랜덤으로 **학습데이터셋(trainning) / 검증데이터셋(validation)으로 분할** 하고 테스트 셋도 같이 준비 합니다. 검증 데이터로는 검증은 시행하지 않고 오직 학습 데이터와 테스트 데이터로만 진행하는 경우가 많습니다. 데이터 분할 비율은 정해진 것이 없습니다.  

일반적으로 학습 데이터는 전체 데이터중 70\~90%, 검증 데이터는 10\~20%, 테스트 데이터는 10\~20% 가량으로 분할 합니다. 또한<u>학습 데이터를 한번에 분석하여 하나의 모델 결과를 얻기 보다 세부 그룹으로 나누어 교차 검증을 하는 방법도 데이터가 적거나 검증 결과를 일반화 시키기 위해 많이 사용</u> 됩니다. (K-FOLD 교차검증)


**2. 데이터 전처리**  
데이터 전처리에서는 **데이터 정규화와 범주형 자료의 원핫인코딩 과정이 핵심**입니다. 다만 특성치(독립변수)에 범주형 자료가 없는 경우 원핫 인코딩은 생략 됩니다.
그리고 원핫 인코딩은 데이터셋을 분할하기 전에 전체 데이터에 적용하는 것이 더 일반적입니다.
정규화는 **표준화(평균 0 / 표준편차 1) 또는 Min-Max 정규화**가 대표적인 표준화 기법입니다.
정규화 과정은 개념만 이해한다면 어렵지 않은 과정입니다.

**3. 모델적용**  
실제 **머신러닝 알고리즘을 데이터에 적용(fit) 하는 단계**입니다. 그 결과 다양한 평가지표를 통해 알고리즘 결과들을 비교하고 최적의 결과를 찾아내야 합니다.
알고리즘이 회귀분석 문제인지, 분류문제인지, 비지도 학습에 해당하는 문제인지에 따라 고려해야 하는 머신 러닝 알고리즘이 다릅니다. 이를 적절히 선택하여 적용하는 것이 핵심입니다.

**4. 하이퍼파라미터튜닝 및 최종 모델 결정**  
최종 분류기에서 검증 셋은 사용하지 않으며, 테스트 셋으로 성능을 평가합니다. 오로지 테스트 데이터의 결과에 집중 하는 단계로 <u>다양한 알고리즘과 다양한 하이퍼파라미터들을 튜닝하면서 최적의 결과를 찾아내는 단계</u> 입니다.


## 1.1 Data Set 분할

전체 분석 데이터 중 머신 러닝 모델 알고리즘을 학습 시키기 위한 학습 데이터(trai data)는 70\~80%, 학습된 모델이 다른 데이터에도 유효한지를 알기 위한 테스트 데이터(test data)를 20\~30% 으로 나눈 것이 일반적이다

이렇게 전체 데이터가 아니라 테스트 데이터를 따로 떼어내어 확인하는 이유는 해당 모델이 다른 데이터에도 유효한 검증력을 가지는지 즉 일반화 가능한 모델인지를 검증하고 확인하기 위한 것이다.

## 1.2 Data 전처리

데이터 셋을 나눈 후, 목적변수(종속변수)인 레이블 (y)을 제외하고 **특성치(독립변수)는 반드시 정규화를 진행하여 데이터를 정제**하여야 합니다.
정규화는 X 변수들의 단위가 제각각이기 때문에 생기는 **스케일 비대칭 문제를 해결하기 위하여 동일한 단위로 정규화 하여 그 변화폭을 동일하게 맞추기 위한 작업**입니다. 또한 머신러닝 알고리즘들은 좌표 공간상의 거리를 중요하게 생각하는 만큼 서로 다른 스케일은 오차 발생의 위험을 높이는 주된 이유이기 때문에 반듯이 정규화 하여 평탄한 스케일을 만들어 주는 작업이 수행 되어야만 합니다.

머실러닝에서 **정규화 작업의 종류에는 표준화(Standardization) 과 민맥스(Min-Max) 정규화** 두가지 방법이 주로 사용됩니다.

또, 특성치(독립변수)에 범주형 자료가 존재하는 경우 이를 그대로 사용하여서는 안되며, 이를 반듯이 원핫인코딩 처리하여 사용하여야 합니다.

ex) 남녀를 1과 2로 분류한 sex 변수가 존재할 경우[남성:0,1] / [여성:0,1] 과 같은 **분류 유형별 0또는 1로 표기하는 방식(원핫인코딩)을 적용**하여 주어야 한다.

## 1.3 모델 학습

데이터 셋을 나누고 특성치 X(독립변수)를 정규화 하여 평탄한 데이터를 만들었다면 이제 머신 러닝 알고리즘을 이용하여 모델을 훈련 시킬 단계 입니다. **목적 변수(종속변수) y가 존재하는 경우에는 지도 학습(supervised learning)을 그렇지 않다면 군집분석 또는 연관분석을 수행**하여야 합니다.

머신러닝 알고리즘은 사이킷런 패키지를 주로 이용합니다. 대부분 알고리즘은 회귀문제(레이블이 연속형), 분류문제(레이블이 범주형) 모두 사용가능합니다. 하지만 <u>로지스틱 회귀 모델 처럼 분류문제에만 적합하거나 선형회귀/릿지/랏소/엘라스틱넷과 같이 호귀 문제에만 적합한 알고리즘</u>도 있습니다.

또한 **여러가지 알고리즘을 병합하여 더 좋은 결과를 도출하는 앙상블(Ensemble)방법**도 존재합니다.

본서에서는 총 18가지 알고리즘을 살펴보고 각 핵심개념과 원리, 주요 하이퍼 파라미터의 의미에 대해 학습하도록 할 것입니다.

## 1.4 하이퍼파라미터 탐색과 모델 튜닝

머신러닝은 다양한 **하이퍼파라미터(hyper-parameter)를 조정하여 데이텅 적합한 최적의 알고리즘을 찾아내는 과정**입니다. 물론 기본 값으로 수행해도 좋은 결과를 가져 올 수도 있지만 보다 좋은 성능을 내기 위해서라면 조정과 적합한 하이퍼파라미터 탐색이 필수 입니다. 이러한 과정을 **하이퍼파라미터 탐색(Search) 또는 모델튜닝(Model Tuning)**이라고 합니다.

하이퍼파라미터의 경우 분석자가 수동으로 조정할수도 있지만, 너무 많은 시간이 소요되는 관계로 <u>사이킷런에서는 그리드 탐색(Grid Search) 또는 랜텀 탐색(Random Search)를 통해 여러 조합의 결과를 파악 할 수 있도록 돕고 있습니다.</u>

**그리드 탐색**은 분석자가 몇 가지 임의의 파라미터를 설정하여 그 결과 중 가장 좋은 정확도(모델평가지표)를 선정 하는 방식입니다.

**랜덤 탐색 방식**은 분석자가 설정한 일정 범위 이내에서 무작위로 하이퍼파라미터를 뽑아 최적의 결과를 도출하는 방식입니다.


하이퍼 파라미터는 규제를 의미하는 일종의 모델 조정계수라고 볼 수 있습니다. 예를 들어, 
서포트벡터머신에서 C와 Gamma 값을 조정하면서 얼마나 분류의 기준을 완화할 것인가를 결정 할 수 있습니다.
또 KNN에서는 이웃수를 조정하면서 얼마나 정교한 분류를 할 것인자를 결정 할 수 있습니다. 하이퍼 파라미터 조정은 해당 데이터 분야에 가장 적합한 값을 찾는 과정입니다.

## 1.5 모델 성능 평가

모델성능(performance)란 궁극적으로 얼마나 정확하게 맞추었는가를 지표화한 것입니다. 분류 문제라면 실제 분류와 예측 분류가 얼마나 일치하였는가 이며, 회귀분석 문제에서는 도출된 예측값과 실제 값이 얼마나 비슷한지를 봐야 합니다.

성능평가에 있어 대표적인 **분석지표로 R-Square 와 오차행렬**이 있습니다.

**오차행렬(confusion matrix)** 은 분류문제에 사용되는 성능평가지표로 실제 분류를 예측치와 결과치를 배치한 교차표로 정분류와 오분류를 쉽게 알수 있도록 나눈 결과표 입니다.오차행렬을 통해 정확도, 민감도, 특이도, 정밀도, F1 Score 등을 산출하여 활용 합니다.

**R-Sqaure(설명력)** 는 회귀 분석 문제에서 사용되는 성능평가지표로 설명력을 의미합니다. R-Square 는 0과 1사이의 범위중 1에 가까울수록 실제값과 예측값의 일치하는 정도가 높은것을 의미합니다. 그외 <u>SSE, AE, MSE, MAE, RMSE, MAPE</u> 가 존재하는데 **RMSE는 실제 값과 예측값의 차이를 평균적으로 계산한 값**입니다.
